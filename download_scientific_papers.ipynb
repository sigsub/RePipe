## Notebook guideing the data extraction process for the scientific paper dataset.
https://www.kaggle.com/datasets/sumitm004/arxiv-scientific-research-papers-dataset/data

```python
import pandas as pd
csv = pd.read_excel('LOCATIOn/OF/YOUR/DATASET/DOWNLOAD')
```

```python
def filter_publish_date(publish_date, cutoff=2010):
    if type(publish_date) == str:
        y = int(publish_date.split('/')[2])
        return y < 26 and y >= cutoff
    else:
        return publish_date.year < 2026 and publish_date.year >= cutoff 

csv = csv[csv['published_date'].apply(lambda x: filter_publish_date(x))]
```

```python
cts = (csv['category'].value_counts())
csv['count'] = csv['category'].apply(lambda x: cts[x])
for c in cts.items():
    print(f"{c[0]}: {c[1]}")
```

```python
counts = (csv['category'].value_counts())
s=0
cats = []
for cat, count in counts.items():
    if count > 50 and count < 150 and '(' not in cat:
        s += count
        print(f"{cat}: {count}")
        cats.append(cat)
print(f"Total: {s}")
```

```python
csv = csv[csv['category'].isin(cats)]
csv['arxiv_id'] = csv['id'].apply(lambda x: x.split('-')[-1])
```

```python
from extracting_papers_html import download_ar5iv_data_from_title, download_ar5iv_data
from tqdm.notebook import tqdm
import os
save_dir_base = 'YOUR/SAVE/DIRECTORY/PATH'
for idx, row in tqdm(csv.iterrows(), total=csv.shape[0]):
    save_dir = f"{save_dir_base}/{row['category']}"
    if not os.path.exists(save_dir):
        os.makedirs(save_dir)
    try:
        download_ar5iv_data(row['arxiv_id'], save_dir, None)
    except Exception as e:
        print(f"Error for {row['title']}: {e}")
```
